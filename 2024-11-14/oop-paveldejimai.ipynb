```python
class Studentas() :
    def __init__(self, vardas, pavarde, gimimo_metai) :
        self.vardas = vardas
        self.pavarde = pavarde
        self.gimimo_metai = gimimo_metai

    def get_vardas(self) :
        return f"Vardas: {self.vardas}"
    
    def get_pavarde(self) :
        return f"Pavardė: {self.pavarde}"
    
    def get_gimimo_metai(self) :
        return f"Gimimo metai: {self.pavarde}"
    
    def set_vardas(self, vardas) :
        if len(vardas) > 0 :
            self.vardas = vardas
        else : 
            raise ValueError("Neįvestas vardas")
        
        return self
    
    def set_pavarde(self, pavarde) :
        if len(pavarde) > 0 :
            self.vardas = pavarde
        else : 
            raise ValueError("Neįvestas vardas")
        
        return self
    
    def set_metai(self, metai) :
        if metai != 2024 :
            self.metai = metai
        else : 
            raise ValueError("Negalima priskirti esamų metų")
        
        return self
    

# studentas = Studentas("Jonas", "Jonaitis", 1981)


class Lektorius(Studentas) :
    def __init__(self, vardas, pavarde, gimimo_metai, slaptazodis) :
        # Reikšmių perdavimas tėvinei klasei
        super().__init__(vardas, pavarde, gimimo_metai)

        # Reikšmių nustatymas esamai klasei
        self.slaptazodis = slaptazodis
   
    def get_slaptazodis(self) :
        return f"Jūsų slaptažodis: {self.slaptazodis}"
    
    def set_slaptazodis(self, slaptazodis) :
        if len(slaptazodis) >= 6 :
            self.slaptazodis = slaptazodis
        else : 
            raise ValueError("Slaptazodis per trumpas")
        
        return self
    

lektorius = Lektorius("ANtanas", "Petraitis", 1975, "12345")

print(lektorius.set_slaptazodis("1234568").get_slaptazodis())
```
